```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
fp0 = pd.read_csv('Desktop/dataset/fp0concat.csv')#importer les databases
fp1 = pd.read_csv('Desktop/dataset/fp1concat.csv')
fp2 = pd.read_csv('Desktop/dataset/fp2concat.csv')
fp3 = pd.read_csv('Desktop/dataset/fp3concat.csv')
fp4 = pd.read_csv('Desktop/dataset/fp4concat.csv')
fp5 = pd.read_csv('Desktop/dataset/fp5concat.csv')
fp6 = pd.read_csv('Desktop/dataset/fp6concat.csv')
fp7 = pd.read_csv('Desktop/dataset/fp7concat.csv')
fp8 = pd.read_csv('Desktop/dataset/fp8concat.csv')
```

```python
# FP0 / PRE_FP -> fp0 codé depuis archives, qq mesures d'unification s'imposent:
fp0.rename(columns={'project  ': 'projectAcronym','call': 'topics'}, inplace = True)
fp0['role'] = "coordinator" #création d'une colonne "role" avec le statut de coordonnateur car projets mono-acteurs
fp0columns = ['theme', 'topic']#suppresion de 2 colonnes
fp0.drop(fp0columns, inplace=True, axis=1)
```

```python
database_cordis = pd.concat([fp0,fp1,fp2,fp3,fp4,fp5,fp6,fp7,fp8])
print(database_cordis.shape)
print(database_cordis.isnull().sum()) #nombre de NaN pour chaque colonne
```

```python
database_cordiscolumns = ['Unnamed: 0', 'id_x', 'activityType', 'endOfParticipation', 'organizationUrl', 'vatNumber', 'contactType', 'contactTitle', 'contactFirstNames', 'contactFunction', 'contactTelephoneNumber', 'contactFaxNumber', 'id', 'id_y', 'status']
database_cordis.drop(database_cordiscolumns, inplace=True, axis=1)
database_cordis = database_cordis[database_cordis['name'].notna()] #supprimer lignes NaN de "name"
print(database_cordis.shape)
print(database_cordis.isnull().sum()) #nombre de NaN pour chaque colonne
```

```python
# enregistrer un fichier csv sans éviter le unmaned_1/ réindexage
database_cordis.to_csv("Desktop/database_cordis_01_07_2021.csv",mode = 'w', index=False)
```
