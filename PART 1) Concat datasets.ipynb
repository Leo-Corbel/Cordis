```python
import pandas as pd
import tqdm
import glob
import warnings
import geocoder
import time
import folium
warnings.filterwarnings('ignore')
```

```python
csv = glob.glob("Desktop/dataset/fp1/*.csv")
xls = glob.glob("Desktop/dataset/fp1/*.xlsx")
```

```python
projets = pd.DataFrame()
participants = pd.DataFrame()
```

```python
for path in csv :
    projets = pd.concat([projets,pd.read_csv(path,sep=";")],axis = 0)
    
for path in xls : 
    participants = pd.concat(([participants,pd.read_excel(path)]),axis = 0)
```

```python
participants = participants.rename(columns={'projectRcn':'rcn'})
```

```python
participants = participants[participants.columns[(participants.notna().sum(axis=0) != 0)]]
projets = projets[projets.columns[(projets.notna().sum(axis=0) !=0)]]
projets["rcn"] = projets.rcn.apply(lambda x : int(x))
projets.rcn[projets.rcn.apply(lambda x : type(x) == str)].apply(lambda x : len(x)).value_counts()
```

```python
p_rcn_str = projets[projets.rcn.apply(lambda x : type(x) == str)][['rcn']]
index_del_rcn = p_rcn_str[p_rcn_str.rcn.apply(lambda x:len(x) > 6)].index
projets = projets.loc[~projets.index.isin(index_del_rcn),:]
projets = projets[projets.rcn.notna()]
projets = projets.loc[projets.rcn !='rcn',:]
projets['rcn'] = projets['rcn'].apply(lambda x : int(x))
resultats = participants.merge(projets,left_on="rcn", right_on="rcn", how="left")
```

```python
RAWMAT = resultats[(resultats['programme'] == 'FP1-RAWMAT 3C')] # filter a column from a data
RAWMAT.to_csv("Desktop/RAWMAT1.csv")
```
